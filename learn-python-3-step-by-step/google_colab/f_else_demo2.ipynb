```python
x = (input('Enter first numbert : '))
y = (input('Enter second numbert : '))
z = (input('Enter third numbert : '))

max = 0

if x > y and x > z:
    print('You are in if-block')
    max = x
elif y > z:
    print('You are in elif-block')
    max = y
else:
    print('You are in else-block')
    max = z
print('Maximum is', max)
print('Thank you!!!1')
```
